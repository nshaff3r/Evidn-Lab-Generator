```python
import numpy as np
import pandas as pd
from time import strftime

#import raw spreadsheet, skip first row with title
lab_222 = pd.read_csv("L222 MTCO2.csv",skiprows=[0])

#Drop NaN and 0 values
lab_222 = lab_222.dropna(subset=['Value'])
lab_222 = lab_222.loc[lab_222["Value"]!=0]

lab_222 = lab_222.reset_index()
lab_222 = lab_222.drop("index", axis = 1)
lab_222
```

```python
#determining what month we want to conduct our calculations from
current_month = strftime("%b")
month_dict = {'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4,
             'May': 5, 'Jun':6, 'Jul': 7, 'Aug': 8,
             'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec':12}
#days_dict = {'Jan':31,'Feb': 'Mar','May','Jul', 'Aug', 'Oct','Dec': 31.}
month_num = month_dict[current_month]
calc_month = month_num - 1
#print(calc_month)

month_sum = 0
month_count = 0

#month average
for i in np.arange(0, lab_222.shape[0]):
    day = lab_222["Date"][i]
    #get the month in the current value of the column
    day_split = day.split('/')
    csv_month = day_split[0]
    csv_month = int(csv_month)
    if csv_month == calc_month:
        #print("here")
        month_sum = month_sum + lab_222['Value'][i]
        month_count = month_count + 1
        
        
print(month_count)
month_average = month_sum/month_count
month_average

#clarify how month average is being calculated
```
